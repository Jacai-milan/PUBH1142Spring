# ***Data Wrangling Questions***

## 1. **How do you import the pandas library in Python?**

You import pandas in python by using the `import` statement.

*For example*: `import` `pandas` as `pd`

## 2. ***How do you read a CSV file named data.csv into a pandas DataFrame?***
In order to read a CSV file named data.csv into a pandas DataFrame, you would have to assign it to a variable under the `read` statement.

*For example*: `pd.read_csv('data.csv')`


## 3. **How do you display the first 5 rows of a DataFrame named df?**
To display the first five rows, you will have to use the `head` statement.

*For example*: `df.head()`


## 4. **How do you find the number of rows and columns in a DataFrame df?**
To find the number of row and columns, you use the `shape` statement.

*For example*: `df.shape`


## 5. **How do you print the names of all the columns in a DataFrame df?**
If you want to find the columns names you can use the `.columns` attribute.

*For example*: `df.columns`

## 6. ***How do you get the summary information of the DataFrame df including data types and non-null values?***
To find the summary information of the DataFrame including data types and non-null values, you'll want to use the `.info` method.

*For example*: `df.info`


##  7. ***How do you select only the rows in DataFrame df where the column A is greater than 5?***
Create a `boolean` condition that checks for rows where the value in column A is greater than 5 and assign it to a variable and then print.

*For Example*: `filtered_df = df[df['A'] > 5]`
`print(filtered_df)`

## 8. ***How do you select the columns A and B from the DataFrame df?***
In order to select columns A and B from the DataFrame, you can use the `iloc` statement or you can just use the column names.

*For example*: `df.iloc['Column_A', 'Column_B']` or `df['Column_A', 'Column_B']`

## 9. ***How do you rename the column A to Alpha in DataFrame df?***
if you want to rename a column you can use the `rename` statement with the `inplace=true` parameter. 

*For example*: `df.rename(columns={'A': 'Alpha'}, inplace=True)`


## 10. ***How do you replace all occurrences of the value 999 in DataFrame df with nan?***
If you want to replace al occurences of the value 999 with nan, then you should use the `replace()` method combined with `numpy.nan`.

*For example*: `df.replace(999, np.nan, inplace=True)`


## 11. ***How do you drop all rows in DataFrame df that contain missing values?***
If you want to drop all the rows that contain missiong values you can use the `dropna()` method.

*For example*: `df = df.dropna()`



## 12. ***How do you fill all missing values in DataFrame df with 0?***
To fill all missing values (NaNs) in a DataFrame df with 0, you can use the `fillna()` method.

*For example*: `df = df.fillna(0)`

## 13. ***How do you calculate the mean of column A in DataFrame df?***
To calculate the mean of column A in a DataFrame df, you can use the `mean()` method.

*For example*: `mean_value = df['A'].mean()`

## 14. ***How do you create a new column C in DataFrame df that is the sum of columns A and B?***
You want to assign column A plus column B in to a new column C.

*For example*: `df['C'] = df['A'] + df['B']`


## 15. ***How do you save the DataFrame df to a CSV file named output.csv? (Find the solution by searching the pandas documentation.)***
To save a pandas DataFrame df to a CSV file named output.csv, you can use the `to_csv()` method provided by pandas.

*For example*: `df.to_csv('output.csv', index=False)`
